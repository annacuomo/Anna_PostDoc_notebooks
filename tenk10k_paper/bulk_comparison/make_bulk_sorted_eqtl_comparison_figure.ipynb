```r
# BLUEPRINT: 5,615 CD14 Mono vs our 5,468 CD14 Mono (Mono_C)
# CEDAR: 2,315  CD14 Mono vs our 5,468 CD14 Mono (Mono_C)
# Schmiedel: 1,941 CD14 Mono vs our 5,468 CD14 Mono (Mono_C)
```

```r
options(repr.plot.width = 8, repr.plot.height = 6) 
df_to_plot = data.frame(dataset=c('a_tenk10k','blueprint','cedar','kasela'), 
                  n_egenes=c(5468,5615,1215,0))
p = ggplot(df_to_plot, aes(x=dataset, y=n_egenes, fill=dataset)) 
# p = p + geom_bar(stat='identity', col=col_classical_mono)
p = p + geom_bar(stat='identity')
p = p + scale_fill_brewer(palette = "Set1")
p = p + theme_classic()
p = p + xlab('Dataset') + ylab('N eGenes (p<1e-6)')
p = p + theme(text = element_text(size=20))
p = p + theme(axis.text.x = element_text(angle = 90, vjust = 0.5, hjust=1))
p + ylim(c(0,10000)) + ggtitle('Classical monocytes')
```

```r
options(repr.plot.width = 8, repr.plot.height = 6) 
df_to_plot = data.frame(dataset=c('a_tenk10k','a_tenk10k','blueprint','blueprint','cedar','cedar','kasela','kasela','schmiedel','schmiedel'),
                        replicated=c('no','yes','no','yes','no','yes','no','yes','no','yes'),
                        n_egenes=c(0,5468,5615-3485,3485,1965-1337,1337,0,0,1941-1528,1528))
# head(df_to_plot)
p = ggplot(df_to_plot, aes(x=dataset, y=n_egenes, fill=replicated)) 
# p = p + geom_bar(stat='identity', col=col_classical_mono)
p = p + geom_bar(stat='identity',position="stack")
# p = p + scale_fill_brewer(palette = "Set1")
p = p + theme_classic()
p = p + xlab('Dataset') + ylab('N eGenes (p<1e-6)')
p = p + theme(text = element_text(size=20))
p = p + theme(axis.text.x = element_text(angle = 90, vjust = 0.5, hjust=1))
p + ylim(c(0,10000)) + ggtitle('Classical monocytes')
```
