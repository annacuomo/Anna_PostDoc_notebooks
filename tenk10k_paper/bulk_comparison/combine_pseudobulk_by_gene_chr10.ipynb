```r
chr='chr10'
```

```r
chr_folder = paste0(outdir,chr,'/')
chr_folder
chr_files = list.files(chr_folder)
length(chr_files)
tail(chr_files)
```

```r
chr_genes = unique(gsub('_.*','',chr_files))
length(chr_genes)
head(chr_genes)
```

```r
chrom_out_folder = paste0(outdir,'all_celltypes/',chr,'/')
for (gene in chr_genes){
    out_file = paste0(chrom_out_folder,gene,'_all_celltypes_pheno_cov.tsv')
    if (!file.exists(out_file)){
        df_list = list()
        files = list.files(chr_folder,gene)
#         print(files)
        for (file in files){
            df = as.data.frame(fread(paste0(chr_folder,file)))
            df_list[[file]] = df
        }
        df_gene = as.data.frame(rbindlist(df_list,fill=TRUE))  
        df_gene$V1 = c()

        fwrite(df_gene,out_file,sep='\t')
    }
}
```
