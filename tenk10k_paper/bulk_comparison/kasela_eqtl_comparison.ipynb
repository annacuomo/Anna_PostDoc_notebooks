```r
suppl_tables_dir = '/directflow/SCCGGroupShare/projects/anncuo/TenK10K_pilot/tenk10k/ms_tables/'
df_common_ct = as.data.frame(fread(paste0(suppl_tables_dir,'all_eqtls_fdr_5pct.tsv')))
head(df_common_ct,2)
```

```r
# All CD4
all_celltypes = unique(df_common_ct$celltype)
cd4_celltypes = all_celltypes[grep('CD4',all_celltypes)]
cd4_celltypes = c(cd4_celltypes,'Treg')
cd4_celltypes
```

```r
df_cd4 = df_common_ct[df_common_ct$celltype %in% cd4_celltypes,]
tenk10k_cd4_egenes = unique(df_cd4$gene)
length(tenk10k_cd4_egenes)
```

```r
length(unique(df_cd4[df_cd4$top_pval < 1e-6,]$gene))
```

```r
kasela_catalogue_file_cd4 = paste0(mydir,'kasela_cd4_QTD000371.cc.tsv.gz')
kasela_df_cd4 <- read.csv(gzfile(kasela_catalogue_file_cd4), sep="\t")
nrow(kasela_df_cd4)
head(kasela_df_cd4)
length(unique(kasela_df_cd4$molecular_trait_id))
length(unique(kasela_df_cd4[kasela_df_cd4$pvalue < 1e-6,'molecular_trait_id']))
```

```r
kasela_egenes = unique(kasela_df_cd4[kasela_df_cd4$pvalue < 1e-6,'gene_id'])
length(kasela_egenes)
tenk10k_egenes = unique(df_cd4[df_cd4$top_pval < 1e-6,]$gene)
length(tenk10k_egenes)
length(kasela_egenes[kasela_egenes %in% tenk10k_egenes])
length(kasela_egenes[kasela_egenes %in% tenk10k_egenes])/length(kasela_egenes)
```
