```r
suppl_tables_dir = '/directflow/SCCGGroupShare/projects/anncuo/TenK10K_pilot/tenk10k/ms_tables/'
df_common_ct = as.data.frame(fread(paste0(suppl_tables_dir,'all_eqtls_fdr_5pct.tsv')))
head(df_common_ct)
```

```r
# Classical monocytes (CD14+) 
```

```r
# TenK10K
df_mono_c = df_common_ct[df_common_ct$celltype == 'CD14_Mono',]
tenk10k_monocyte_c_egenes = unique(df_mono_c$gene)
length(tenk10k_monocyte_c_egenes)
length(unique(df_mono_c[df_mono_c$top_pval < 1e-6,]$gene))
```

```r
# Schmiedel et al 
schmiedel_catalogue_file_mono_c = paste0(mydir,'schmiedel_cd14_mono_QTD000504.cc.tsv.gz')
schmiedel_df_mono_c <- read.csv(gzfile(schmiedel_catalogue_file_mono_c), sep="\t")
nrow(schmiedel_df_mono_c)
head(schmiedel_df_mono_c,2)
```

```r
length(unique(schmiedel_df_mono_c$molecular_trait_id))
length(unique(schmiedel_df_mono_c[schmiedel_df_mono_c$pvalue < 1e-6,'molecular_trait_id']))
```

```r
schmiedel_egenes = unique(schmiedel_df_mono_c[schmiedel_df_mono_c$pvalue < 1e-6,'molecular_trait_id'])
length(schmiedel_egenes)
tenk10k_egenes = unique(df_mono_c[df_mono_c$top_pval < 1e-6,]$gene)
length(tenk10k_egenes)
length(schmiedel_egenes[schmiedel_egenes %in% tenk10k_egenes])
length(schmiedel_egenes[schmiedel_egenes %in% tenk10k_egenes])/length(schmiedel_egenes)
```
