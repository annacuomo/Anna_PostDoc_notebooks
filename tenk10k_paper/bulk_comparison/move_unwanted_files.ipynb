```python
import os
import shutil
```

```python
import pandas as pd
```

```python
# get list of genes to exclude
saige_dir = '/directflow/SCCGGroupShare/projects/anncuo/TenK10K_pilot/tenk10k/eqtl_results/saige_qtl/december24_freeze/'
celltypes = os.listdir(saige_dir)
exclude = {'annotate', 'coloc_results', 'README.md', 'CD4_TCM_sample_perm0'}
celltypes = [f for f in celltypes if f not in exclude]
len(celltypes)
```

```python
included_genes = set()
for celltype in celltypes:
    common_acat_file = f'{saige_dir}{celltype}/{celltype}_all_cis_cv_gene_level_results.tsv'
    common_acat_df = pd.read_csv(common_acat_file, sep='\t')
    genes = common_acat_df['gene'].dropna().astype(str).unique()
    included_genes.update(genes)
```

```python
print(f"Collected {len(included_genes)} included genes.")
```

```python
for i in range(22):
    source_folder = f"/directflow/SCCGGroupShare/projects/anncuo/TenK10K_pilot/tenk10k/review_files/all_celltypes_pseudobulk/all_celltypes/chr{i+1}/"
    target_folder = f"/directflow/SCCGGroupShare/projects/anncuo/TenK10K_pilot/tenk10k/review_files/all_celltypes_pseudobulk/all_celltypes/excluded_files/chr{i+1}/"
    os.makedirs(target_folder, exist_ok=True)
    for filename in os.listdir(source_folder):
        filepath = os.path.join(source_folder, filename)

        # Skip directories
        if not os.path.isfile(filepath):
            continue

        # Match only files that follow the pattern: {gene}_all_celltypes_pheno_cov.tsv
        if not filename.endswith("_all_celltypes_pheno_cov.tsv"):
            continue

        gene_name = filename.replace("_all_celltypes_pheno_cov.tsv", "")

        if gene_name not in included_genes:
            shutil.move(filepath, os.path.join(target_folder, filename))
            print(f"Moved excluded file: {filename}")
```
