```r
blueprint_df <- fread(paste0(mydir,'blueprint_1-s2.0-S0092867416314465-mmc3.csv'))
colnames(blueprint_df) = as.character(blueprint_df[4,])
colnames(blueprint_df)[4:13] = paste0('Monocyte_',colnames(blueprint_df)[4:13])
colnames(blueprint_df)[14:23] = paste0('Neutrophil_',colnames(blueprint_df)[14:23])
colnames(blueprint_df)[24:30] = paste0('CD4_',colnames(blueprint_df)[24:30])
blueprint_df <- blueprint_df[-c(1:4),]
nrow(blueprint_df)
head(blueprint_df,2)
```

```r
monocyte_df = blueprint_df[blueprint_df$Monocyte_eQTL == 1,]
nrow(monocyte_df)
head(monocyte_df)
```

```r
monocytes_egenes = gsub('\\..*','',unique(monocyte_df[['ENSEMBL ID']]))
length(monocytes_egenes)
head(monocytes_egenes)
```
