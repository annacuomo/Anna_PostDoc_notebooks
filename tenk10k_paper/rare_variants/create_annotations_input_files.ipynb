```r
saige_dir = '/directflow/SCCGGroupShare/projects/anncuo/TenK10K_pilot/tenk10k/eqtl_results/saige_qtl/december24_freeze/'
celltypes = list.files(saige_dir)
celltypes = celltypes[!(celltypes %in% c('annotate','coloc_results','README.md','CD4_TCM_sample_perm0'))]
length(celltypes)
celltypes
```

```r
genes_tested = c()
for (celltype in celltypes){
    common_acat_file = paste0(saige_dir,celltype,'/',celltype,'_all_cis_rv_set_test_results.tsv')
    common_acat_df = as.data.frame(fread(common_acat_file))
    genes_tested = unique(c(genes_tested,unique(common_acat_df$Region)))
}
length(genes_tested)
head(genes_tested)
```

```r
celltype = 'CD4_TCM'
rare_sv_file = paste0(saige_dir,celltype,'/',celltype,'_rare_all_cis_raw_pvalues.tsv')
rare_sv_df = as.data.frame(fread(rare_sv_file))
rare_sv_df$POS = as.numeric(rare_sv_df$POS)
nrow(rare_sv_df)
head(rare_sv_df)
```

```r
df_gene_cis = list()
for (gene in unique(rare_sv_df$gene)){
    df0 = rare_sv_df[rare_sv_df$gene == gene,]
    chr = paste0('chr',df0$CHR[1])
    start = min(df0$POS, na.rm=TRUE)
    end = max(df0$POS, na.rm=TRUE)
    df_gene_cis[[gene]] = data.frame(chr=chr,start=start,end=end)
}
df_genes = as.data.frame(rbindlist(df_gene_cis))
head(df_genes)
```
